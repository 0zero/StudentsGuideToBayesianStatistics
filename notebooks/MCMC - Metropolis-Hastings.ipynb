```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import seaborn as sns
```

### Problem 13.1 Ticked off

```python
alpha = 1
beta = 1
sample_size = 100
number_of_samples = 1000
```

```python
def prior_predictive(num_samples, a, b, n):
    samples = np.zeros(num_samples)
    for i in range(0,num_samples):
        theta = np.random.beta(a,b,size=1)
        samples[i] = np.random.binomial(n=n,p=theta)
    return samples
```

```python
prior_pred = prior_predictive(
    num_samples=number_of_samples, 
    a=alpha,
    b=beta, 
    n=sample_size
)
```

```python
plt.hist(prior_pred);
```

beta prior and binomial sample/likelihood gives a beta-binomial posterior via conjugate priors

```python
beta_bin_mean = sample_size * (alpha/(alpha + beta))

print(f"Prior predictive mean is {prior_pred.mean()} which should be similar", 
      f"to the theoretical beta-binomial mean of {beta_bin_mean}.")
```

Posterior distribuution is Beta(a+ sumX, b + N - sumX)

```python
ticks_in_single_sample = 6
a = alpha + ticks_in_single_sample
b =  beta + sample_size - ticks_in_single_sample
```

```python
posterior_mean = a/(b+a)
print(f"The posterior mean is {posterior_mean}.")
```

```python
posterior = np.random.beta(a,b,sample_size)
```

```python
sns.distplot(posterior);
sns.distplot(np.random.beta(a,b,10000),hist=False)
plt.xlim([0,.2]);
```

```python
def get_means(a, b, min_sample_size, max_sample_size):
    sample_sizes = np.arange(min_sample_size, max_sample_size,step=10)
    means = np.zeros(len(sample_sizes))
    variances = np.zeros(len(sample_sizes))
    for i in range(0,len(sample_sizes)):
        means[i] = np.random.beta(a,b,sample_sizes[i]).mean()
        variances[i] = np.random.beta(a,b,sample_sizes[i]).var()
    return (sample_sizes, means, variances)
```

```python
samples,means,variances = get_means(a,b,100,1000)
```

```python
plt.plot(samples,means)
plt.plot(plt.xlim(),[posterior_mean,posterior_mean])
```
