```python
import arviz as az
import numpy as np
import pymc3 as pm
import matplotlib.pyplot as plt
```

```python
y=np.random.normal(1.5,0.2,10)
```

```python
with pm.Model() as height_model:
    mu = pm.Normal("mu",mu=1.5,sigma=0.1)
    sigma = pm.Gamma("sigma",1,1)
    
    y_likelihood = pm.Normal("y_likelihood",mu,sigma,observed=y)
```

```python
with height_model:
    trace = pm.sample(draws=200,tune=100,chains=4)
    az.plot_forest(trace, var_names=['mu', 'sigma'], r_hat=True)
```

```python
az.plot_trace(trace);
```

```python
az.summary(trace, round_to=2)
```
